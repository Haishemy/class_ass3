```python
import pandas as pd
import numpy as np
```

```python
dataset= pd.read_csv('updated_pollution_dataset.csv')
dataset
```

```python
from sklearn.preprocessing import OrdinalEncoder
np.random.seed(0)

data = {
    'Temperature': np.random.uniform(20, 40, size=5000), 
    'Humidity': np.random.uniform(30, 100, size=5000),     
    'PM2.5': np.random.uniform(0, 150, size=5000),         
    'PM10': np.random.uniform(0, 150, size=5000),         
    'NO2': np.random.uniform(0, 50, size=5000),            
    'SO2': np.random.uniform(0, 30, size=5000),           
    'CO': np.random.uniform(0, 5, size=5000),              
    'Proximity_to_Industrial_Areas': np.random.uniform(0, 20, size=5000), 
    'Population_Density': np.random.randint(100, 1000, size=5000),         
    'Air Quality': np.random.choice(['Good', 'Moderate', 'Hazardous'], size=5000)  
}

dataset = pd.DataFrame(data)

encoder = OrdinalEncoder(categories=[['Good', 'Moderate', 'Hazardous']])

dataset['Air Quality Encoded'] = encoder.fit_transform(dataset[['Air Quality']])

print(dataset.head())  
```

```python
dataset
```

```python
dataset=dataset.drop("Air Quality",axis=1)
dataset
```

```python
dataset.columns
```

```python
indep= dataset[['Temperature', 'Humidity', 'PM2.5', 'PM10', 'NO2', 'SO2', 'CO',
       'Proximity_to_Industrial_Areas', 'Population_Density']]
dep= dataset[['Air Quality Encoded']]
```

```python
dataset['Air Quality Encoded'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(indep, dep, test_size=0.30,random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier= DecisionTreeClassifier()
classifier.fit(x_train,y_train)
```

```python
y_pred= classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
con_mat= confusion_matrix(y_test,y_pred)
con_mat
```

```python
print(con_mat)
```

```python
from sklearn.metrics import classification_report
class_report=classification_report(y_test,y_pred)
print(class_report)
```
